```python
import pandas as pd
data = pd.read_csv('bj.csv')
```

```python
data.describe()
```

```python
data = data.drop(['NO2_Concentration', 'CO_Concentration', 'SO2_Concentration'], axis=1)
```

```python
data.head()
```

```python
data.set_index(['station_id'], inplace=True)
```

```python
station_ids = data.index.unique().values
```

```python
every_station_data = []
for station in station_ids:
    every_station_data.append(data.loc[station,:])
```

```python
every_station_data[0]
```

```python
every_station_data[8]
```

```python
for d in every_station_data:
    d.dropna(0, 'any', inplace=True)
```

```python
every_station_data[1].isnull().sum()
```

```python
for i, d in enumerate(every_station_data):
    print(station_ids[i], len(d))
    if len(d) < 240:
        print("Falut on", station_ids[i]) # 8 沒救
        print(d)
```

```python
for i in range(len(every_station_data)):
    every_station_data[i] = every_station_data[i].iloc[-240:]
```

```python
every_station_data[8]
```

```python
encode_csv = pd.concat(every_station_data)
```
