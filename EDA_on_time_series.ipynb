```python
conf_dataset = dataset.dropna(axis=0, how='any')
```

```python
d = pd.to_datetime(conf_dataset['utc_time']).dt
```

```python
conf_dataset['hour'] = d.hour
conf_dataset['day'] = d.day
conf_dataset['month'] = d.month
conf_dataset['year'] = d.year
conf_dataset['dayofweek'] = d.dayofweek
conf_dataset['dayofyear'] = d.dayofyear
```

```python
conf_dataset.drop(['utc_time'], axis=1).values
```

```python
conf_dataset['is_weekend'] = (conf_dataset['dayofweek']==5) | (conf_dataset['dayofweek']==6)
conf_dataset['work_time'] = (conf_dataset.hour > 6) & (conf_dataset.hour < 20)
 
```

```python
each_day = [conf_dataset[conf_dataset.dayofweek == i] for i in range(0, 7)]
```

```python
for day in each_day:
    print(day['PM2.5'].mean())
    print(day['PM10'].mean())
    print(day['NO2'].mean())
    print(day['CO'].mean())
    print(day['O3'].mean())
    print(day['SO2'].mean())
    print('\n')
```
